```python
#Import estimated vol by Garch(1,1)
#Import mean return by Garch(1,1)
Garch=pd.read_csv("8K data/Garch_results.csv")
X_ts=pd.read_csv("8K data/X1.csv",index_col=0)
X_text=pd.read_csv("8K data/X2.csv",index_col=0)
Return=pd.read_csv("8K data/Y.csv",index_col=0)
#Garch_mu=Garch["mu"]
#Garch_vol=Garch["vol"]
```

```python
Garch.index=df.index
X_ts.index=df.index
X_text.index=df.index
Return.index=df.index
```

```python
X_ts.columns=index_lag[0:100]
X_text.columns=df_temp.columns
```

```python
Garch_reg=Garch[Garch["Vol"]!="Error : Exiting\r\n"]
```

```python
Garch_reg["mu"]=Garch_reg["mu"].apply(lambda x:float(x))
```

```python
Garch_reg["Vol"]=Garch_reg["Vol"].apply(lambda x:float(x))
```

```python
X_ts_reg=X_ts.loc[Garch_reg.index].copy()
X_text_reg=X_text.loc[Garch_reg.index].copy()
Return_reg=Return.loc[Garch_reg.index].copy()
```

```python
Abs_return=np.abs((Return_reg["Return"]-Garch_reg["mu"]))
```

```python
Abs_return=pd.DataFrame(Abs_return)
Abs_return.columns=["Abs_return"]
Abs_return.shape
```

## Lasso Regression - to update

### Text Only

```python
#construct model and predict
t_sample=700
best_param=TA.Lasso_param(X_text_reg,Abs_return, 0.7)
lasso =linear_model.Lasso(alpha=best_param)
lasso.fit(X_text_reg.iloc[:t_sample,:],Abs_return.iloc[:t_sample])
pred=lasso.predict(X_text_reg.iloc[:t_sample,:])
```

```python
print(lasso.coef_)
print(lasso.intercept_)
print(pred)
```

```python
Comp=pd.DataFrame({"Abs_return":Abs_return.iloc[t_sample:,0],
                  "Pred_vol":pred})
Comp["error"]=Comp["Abs_return"]-Comp["Pred_vol"]
#MSE=1/Comp.shape[0]*(sum(Comp["error"]^2))
```

### Text+Garch

```python
X_all_reg=X_text_reg.copy()
X_all_reg["Garch_vol"]=Garch_reg["Vol"]
X_all_reg.shape
```

```python
#construct model and predict
best_param2=TA.Lasso_param(X_all_reg,Abs_return, 0.7)
lasso =linear_model.Lasso(alpha=best_param2)
lasso.fit(X_all_reg.iloc[:t_sample,:],Abs_return.iloc[:t_sample])
pred=lasso.predict(X_all_reg.iloc[t_sample:,:])
```

```python
print(lasso.coef_)
print(lasso.intercept_)
print(pred)
```

## SVR Regression - to update 

## Text Only#

```python
#To update
best_power=TA.SVR_param(X_text_reg,Abs_return, 0.7) 
```

```python
#construct model and predict
t_sample=700
svr = SVR(kernel='linear',C=np.exp(best_power),epsilon=0.01)
svr.fit(X_text_reg.iloc[:t_sample,:],Abs_return.iloc[:t_sample])
pred=svr.predict(X_text_reg.iloc[t_sample:,:])
```

```python
print(svr.coef_)
print(svr.intercept_)
print(pred)
```

```python
Comp=pd.DataFrame({"Abs_return":Abs_return.iloc[t_sample:,0],
                  "Pred_vol":pred})
Comp["Error"]=Comp["Abs_return"]-Comp["Pred_vol"]
```

```python
Comp["Garch_vol"]=Garch.iloc[t_sample:,0]
Comp
```
